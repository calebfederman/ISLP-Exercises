```python
# 8.d

filter_columns = ['Top10perc','Apps','Enroll']

pd.plotting.scatter_matrix(college[filter_columns]);
```

```python
# 8.f

college['Elite'] = pd.cut(college['Top10perc']/college['Enroll'],
                          [0,0.5,1],
                          labels=['No','Yes'])

college['Elite'].value_counts()
```

```python
college.boxplot('Outstate', by='Elite');
```

```python
# 8.g

fig, ax = plt.subplots(2, 2)
college.hist('PhD',bins=20,ax=ax[0][0],grid=False)
college.hist('Grad.Rate',bins=20,ax=ax[0][1],grid=False)
college.hist('Outstate',bins=5,ax=ax[1][0],grid=False)
college.hist('Room.Board',bins=5,ax=ax[1][1],grid=False);
```

```python
# 8.h

college.describe()
```

```python
fig, ax = plt.subplots(1,3, figsize=(9,3))
ax[0].set_xlabel('% of Faculty with PhD')
ax[0].set_ylabel('Graduation Rate')
ax[0].scatter(college['PhD'],college['Grad.Rate'], marker='.')
ax[1].set_xlabel('Student Faculty Ratio')
ax[1].set_ylabel('Graduation Rate')
ax[1].scatter(college['S.F.Ratio'],college['Grad.Rate'], marker='.')
ax[2].set_xlabel('Expenditure per Student')
ax[2].set_ylabel('Graduation Rate')
ax[2].scatter(college['Expend'],college['Grad.Rate'], marker='.');
```
